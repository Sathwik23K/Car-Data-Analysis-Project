1) For Data Cleaning
# **Find all the null values and then fill it with the mean of that column**

```python
car.isnull().sum()
```

```python
car['Cylinders'].fillna(car['Cylinders'].mean(), inplace=True)
```

```python
# Calculate the mean for each numeric column
numeric_means = car.select_dtypes(include='number').mean()

# Fill NaN values in numeric columns with their respective mean
car.fillna(numeric_means, inplace=True)
# Calculate the mode for each non-numeric column
non_numeric_modes = car.select_dtypes(exclude='number').mode().iloc[0]

# Fill NaN values in non-numeric columns with their respective mode
car.fillna(non_numeric_modes, inplace=True)

# Now fill NaN values in numeric columns with their respective mean
numeric_means = car.select_dtypes(include='number').mean()
car.fillna(numeric_means, inplace=True)
```
